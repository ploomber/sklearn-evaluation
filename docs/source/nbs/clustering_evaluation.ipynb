# Clustering evaluation

Learn how to easily evaluate clusters.

- Determine optimal number of clusters using Elbow curve.
- Determine if individual points are correctly assigned to their clusters using Silhouette analysis

```python
import matplotlib
from sklearn import datasets
from sklearn.cluster import KMeans
from sklearn_evaluation import plot
```

```python
matplotlib.rcParams["figure.figsize"] = (7, 7)
matplotlib.rcParams["font.size"] = 18
```

```python
# get data for clustering
X, y = datasets.make_blobs(
    n_samples=500,
    n_features=2,
    centers=4,
    cluster_std=1,
    center_box=(-10.0, 10.0),
    shuffle=True,
    random_state=1,
)

# Fit kMeans on the data
kmeans = KMeans(random_state=10)
```

### Silhouette plot for default range_n_clusters (2,3,4,5,6)

##### This plot shows that n_clusters value of 3, 5 and 6 are a bad pick for the given data. One is likely to select between 2 and 4 n_clusters.

```python
silhouette = plot.silhouette_plot(X, kmeans)
```

### Silhouette plot from cluster labels

```python
X, y = datasets.make_blobs(
    n_samples=500,
    n_features=2,
    centers=4,
    cluster_std=1,
    center_box=(-10.0, 10.0),
    shuffle=True,
    random_state=1,
)

kmeans = KMeans(n_clusters=4, random_state=1)
cluster_labels = kmeans.fit_predict(X)
```

```python
silhouette = plot.silhouette_plot_from_results(X, cluster_labels)
```
